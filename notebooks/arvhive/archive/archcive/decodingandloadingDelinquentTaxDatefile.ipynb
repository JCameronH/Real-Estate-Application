```python
import pandas as pd

path = r"C:\Users\howar\OneDrive\Documents\BU Mod2\699 pt2\Real Estate\All Data Aggregated in CSV\Delinquint Tax Data HCAD\Hctax_DelTaxDataFile_March2026.txt"

records = []

with open(path, "r", encoding="latin-1", errors="replace") as f:
    for line in f:
        line = line.strip()

        # remove leading control characters
        if not line[0].isdigit():
            line = line[1:]

        acct = line[0:13]
        tax_year = line[13:17]
        status = line[17:18]
        rest = line[18:]

        records.append({
            "acct": acct,
            "tax_year": tax_year,
            "status": status,
            "raw_balance_data": rest
        })

df = pd.DataFrame(records)

print(df.head(10))
print(df.shape)
```

```python
records = []

with open(path, "r", encoding="latin-1", errors="replace") as f:
    for line in f:
        line = line.rstrip("\n")

        flag = line[1:2]
        acct = line[2:15]
        tax_year = line[15:19]
        aflag = line[19:20]
        exemption = line[278:280]

        records.append({
            "acct": acct.strip(),
            "tax_year": int(tax_year),
            "flag": flag.strip(),
            "aflag": aflag.strip(),
            "exemption_code": exemption.strip()
        })

df = pd.DataFrame(records)
print(df.head())
```

```python
df["flag"] = df["flag"].astype("string").str.strip()
df["aflag"] = df["aflag"].astype("string").str.strip()
df["exemption_code"] = df["exemption_code"].astype("string").str.strip()
```

```python
df["is_delinquent_record"] = True  # this file is delinquency-history, so each row is a delinquent-year record
df["is_payment_plan"] = df["aflag"].eq("P")
df["is_judgment"] = df["aflag"].eq("J")
df["is_suit"] = df["aflag"].eq("S")
df["is_bankruptcy"] = df["aflag"].eq("Y")
df["is_tax_deferred"] = df["aflag"].eq("X")
```

```python
features_tax = (
    df.groupby("acct", as_index=False)
      .agg(
          delinquent_years=("tax_year", "nunique"),
          first_delinquent_year=("tax_year", "min"),
          most_recent_delinquent_year=("tax_year", "max"),
          payment_plan_years=("is_payment_plan", "sum"),
          judgment_years=("is_judgment", "sum"),
          suit_years=("is_suit", "sum"),
          bankruptcy_years=("is_bankruptcy", "sum"),
          deferred_years=("is_tax_deferred", "sum"),
      )
)
print(features_tax.head())
```
